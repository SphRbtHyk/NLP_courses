```python
from typing import Union
```

```python
def mystere(x: Union[list[int], list[float]]) -> Union[int, float]:
    return sum(x)
```

```python
def mystere(y: Union[list[int], list[float]]) -> Union[list[int], list[float]]:
    return [x**2 for x in y if x % 3 == 0]
```

```python
mystere([3, 2, 10])
```

```python
def mystere(x: list[int]) -> int:
    try:
        return x[2]
    except IndexError:
        return None
        #raise IndexError(f"Pas possible d'accéder au 2ème élément car la longueur de la liste est {len(x)}")
```

```python
mystere([2])
```

```python
def mystere(y: dict[str, int]) -> list[str]:
    message = []
    for x in y.values():
        if x == 3:
            message.append("ok")
        else:
            message.append("not ok")
    return message
```

```python
def mystere(x: dict[str, str]) -> dict[str, str]:
    demo = {}
    for ix, jx in x.items():
        print(f"{jx}-test")
        demo[f"{jx}-test"] = ix
        print("---")
    return demo
```

```python
mystere({"1": "1", "2": "1", "3": "4"})
```

```python
def mystere(v: dict[str, float]) -> dict[str, float]:
    return {f"{k}-demo": l**2 for k, l in v.items() if l % 3 != 0}
```
