### Similarité cosinus

```python
from math import sqrt

def norm_euclidean(liste: list[float]) -> float:
    """Compute the euclidean norm.

    Args:
        liste (list[float]): Input vector.

    Returns:
        float: Value of the norm.
    """
    norm = 0
    for elem in liste:
        norm += elem ** 2
    return sqrt(norm)
    # return sqrt(sum([elem**2 for elem in liste]))
```

```python
def cosine_similarity(liste_1: list[float], liste_2: list[float]) -> float:
    """Computes the cosine similarity between two vectors.

    Args:
        liste_1 (list[float]): Vector 1
        liste_2 (list[float]): Vector 2
    """
    scalar_product = 0
    for val_1, val_2 in zip(liste_1, liste_2):
        scalar_product += val_1 * val_2    

    return scalar_product / ((norm_euclidean(liste_1)*norm_euclidean(liste_2)))

cosine_similarity([1/5, 0, 1/5, 0, 1/5], [0, 1/5, 1/5, 1/5, 0])
```

```python
from math import sqrt

def euclidean_distance(liste_1, liste_2):
    distance = 0
    for val_1, val_2 in zip(liste_1, liste_2):
        distance += (val_1 - val_2)**2
    return sqrt(distance)
```

```python
def euclidean_distance(liste_1: list[float],
                       liste_2: list[float]) -> float:
    return sqrt(sum([(elem1 - elem2)**2 for elem1, elem2 in zip(liste_1, liste_2)]))
```

```python
def cosine_similarity(liste_1, liste_2):
    scalar_product = 0
    for val_1, val_2 in zip(liste_1, liste_2):
        scalar_product += val_1 * val_2
    return scalar_product/((euclidean_distance(liste_1, liste_1))*(euclidean_distance(liste_2, liste_2)))
```
