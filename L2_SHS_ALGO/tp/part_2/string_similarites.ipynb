# Similarité entre chaîne de caractères

Étant donné le pseudo-code distribué en cours, implémentez:

1. La similarité de Hamming
2. La similarité de Jaro et de Jaro-Winkler
3. La similarité cosinus
4. La similarité Jaccard

Pour chacune des similarités, vous préciserez sa complexité et vous réaliserez une évaluation empiring à l'aide d'un *benchmarking* réalisé grâce à la commande `time`.
